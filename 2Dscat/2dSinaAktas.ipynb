```python
ang=[20,40,60,80,100,120,140,160,180]
dn=[42,48,63,76,107,111,137,162,172]

%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from scipy.odr import *



xx= np.round(np.sin(np.radians(np.array(ang))/2),2)
yer= np.sqrt(np.array(dn))

def fit_func(p, x):
    m ,c= p
    return m*x+c

linear = Model(fit_func)

data = RealData(xx, dn, sy=yer)

odr= ODR(data, linear, beta0=[0., 1.])

out = odr.run()

out.pprint()

x_fit = np.linspace(xx[0]-0.1, xx[-1]+0.1, 1000)
y_fit = fit_func(out.beta, x_fit)

plt.errorbar(xx, dn, yerr=yer, linestyle='None', marker='x',ecolor='r')

plt.plot(x_fit, y_fit)


plt.xlabel('sin(Î¸/2)')
plt.ylabel('dN')

plt.show()
```
