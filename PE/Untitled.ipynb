```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from scipy.odr import *
from scipy.stats import linregress
import random



v=[0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5]
i=[1.6e-13, 1.2e-13, 8e-14, 6e-14, 4e-14, 2.6e-14, 1.4e-14, 2e-15, -6e-15, -2e-14]
ve=[0.001, 0.001, 0.001, 0.001, 0.001, 0.001, 0.001, 0.001, 0.001, 0.001]
ie=[2e-14, 2e-14, 2e-14, 2e-14, 2e-14, 2e-15, 2e-15, 2e-15, 2e-15, 2e-14]


mid=8

vp=v[:8]
ip=i[:8]

vne=v[8:]
ine=i[8:]


fit = np.polyfit(v[:mid], i[:mid], 1)  
yp1f = np.poly1d(fit)

fit2=np.polyfit(v[mid:], i[mid:], 1) 

yn1f = np.poly1d(fit2)

plt.plot(v[mid:], yn1f(v[mid:]), '-')
plt.plot(v[:mid], yp1f(v[:mid]), '-')



# Define a function (quadratic in our case) to fit the data with.
def fit_func(p, x):
    m, c = p
    return m*x + c

# Create a model for fitting.
fit_model = Model(fit_func)

# Create a RealData object using our initiated data from above.
data1 = RealData(v[:mid], i[:mid], sx=ve[:mid], sy=ie[:mid])
data2 = RealData(v[mid:], i[mid:], sx=ve[mid:], sy=ie[mid:])
# Set up ODR with the model and data.
odr1 = ODR(data1, fit_model, beta0=[0., 1.])
odr2 = ODR(data2, fit_model, beta0=[0., 1.])

# Run the regression.
out1 = odr1.run()
out2 = odr2.run()

# Use the in-built pprint method to give us results.
out1.pprint()
out2.pprint()
'''Beta: [ 1.01781493  0.48498006]
Beta Std Error: [ 0.00390799  0.03660941]
Beta Covariance: [[ 0.00241322 -0.01420883]
 [-0.01420883  0.21177597]]
Residual Variance: 0.00632861634898189
Inverse Condition #: 0.4195196193536024
Reason(s) for Halting:
  Sum of squares convergence'''

xp_fit = np.linspace(v[0], v[mid-1], 1000)
xn_fit = np.linspace(i[mid], i[-1], 1000)
yp_fit = fit_func(out1.beta, xp_fit)
yn_fit = fit_func(out2.beta, xn_fit)

plt.errorbar(v, i, xerr=ve, yerr=ie, linestyle='None', marker='x',ecolor='r')
plt.plot(xp_fit, yp_fit)
plt.plot(xn_fit, yn_fit)

plt.show()


slope, intercept, r_value, p_value, std_err1 = linregress(v[mid:],i[mid:])
print(slope)
print(r_value)
print(p_value)
print(std_err1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.odr import *

import random





# Initiate some data, giving some randomness using random.random().
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([i**2 + random.random() for i in x])
z = np.array([0, 1.3, 2.1, 2.8, 4, 5.8])

x_err = np.array([random.random() for i in x])
y_err = np.array([random.random() for i in x])

# Define a function (quadratic in our case) to fit the data with.
def quad_func(p, x):
    m, c = p
    return m*x + c

# Create a model for fitting.
quad_model = Model(quad_func)

# Create a RealData object using our initiated data from above.
data = RealData(x, z, sx=x_err, sy=y_err)

# Set up ODR with the model and data.
odr = ODR(data, quad_model, beta0=[0., 1.])

# Run the regression.
out = odr.run()

# Use the in-built pprint method to give us results.
out.pprint()
'''Beta: [ 1.01781493  0.48498006]
Beta Std Error: [ 0.00390799  0.03660941]
Beta Covariance: [[ 0.00241322 -0.01420883]
 [-0.01420883  0.21177597]]
Residual Variance: 0.00632861634898189
Inverse Condition #: 0.4195196193536024
Reason(s) for Halting:
  Sum of squares convergence'''

x_fit = np.linspace(x[0], x[-1], 1000)
y_fit = quad_func(out.beta, x_fit)

plt.errorbar(x, z, xerr=x_err, yerr=y_err, linestyle='None', marker='x')
plt.plot(x_fit, y_fit)

plt.show()

```
