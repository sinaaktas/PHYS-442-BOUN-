```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from scipy.odr import *
from scipy.stats import linregress
from scipy.optimize import fsolve
import random




import pylab
import pandas

colnames = ['A1', 'B1']
data = pandas.read_excel("15_kv.xlsx",  names=colnames)
icc = data.A1.tolist()
vcc = data.B1.tolist()
angle=icc[58:63]  # arrange where we get th data
intens=vcc[58:63]
angle=np.array(angle)
intens=np.array(intens)


n = len(angle)                       #the number of data
mean = sum(angle)/n                   #note this correction
sigma = np.sqrt(sum((angle-mean)**2)/n)       #note this correction


def fit_func(p, x):
    m ,c= p
    return m*x+c

linear = Model(fit_func)

data = RealData(angle, intens, sy=np.sqrt(angle))

odr= ODR(data, linear, beta0=[0., -1.])

out = odr.run()

out.pprint()

x_fit = np.linspace(angle[0]-0.5, angle[-1]+0.5, 1000)
y_fit = fit_func(out.beta, x_fit)



plt.errorbar(angle, intens, xerr=0, yerr=np.sqrt(intens), linestyle='None', marker='x',ecolor='r')
plt.xlabel('Angle Theta[Degrees]')
plt.ylabel('Intensity of reflected X-ray [Imp/s]')
plt.plot(icc[:100], vcc[:100],'b.')
plt.plot(x_fit,y_fit,'g-')


plt.show()

print('intercection of x-line[degree]',-out.beta[1]/out.beta[0])
print('sigma[degree]',np.sqrt((out.sd_beta[1]/out.beta[0])**2+((out.beta[1]*out.sd_beta[0])/out.beta[0]**2)**2))
print('intercection of x-line[radian]',-(np.deg2rad(out.beta[1])/out.beta[0]))
print('sigma[radian]',np.sqrt((np.deg2rad(out.sd_beta[1])/out.beta[0])**2+((np.deg2rad(out.beta[1])*out.sd_beta[0])/out.beta[0]**2)**2))
```
